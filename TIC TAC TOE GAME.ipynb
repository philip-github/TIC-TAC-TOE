```python
from IPython.display import clear_output
import random

'''
this is a function(display_board) will take a list and display it as X-O table
'''

def display_board(board):
    
    # this is a method to clear the old output for this function...
    
    clear_output()
    
    print("        |        |        ")
    print(f"{board[7]: ^8s}|{board[8]: ^8s}|{board[9]: ^8s}")
    print("        |        |        ")
    print("--------------------------")
    print("        |        |        ")
    print(f"{board[4]: ^8s}|{board[5]: ^8s}|{board[6]: ^8s}")
    print("        |        |        ")
    print("--------------------------")
    print("        |        |        ")
    print(f"{board[1]: ^8s}|{board[2]: ^8s}|{board[3]: ^8s}")
    print("        |        |        ")

'''
this is a function to chooss marker X or O
'''

def player_input():
    marker = ''
    
    # this is a while loop to keep iterate an input if the marker value not X or O...
    
    while marker != "X" and marker != "O":
        marker = input("Please choose X or O: ").upper()
        
    # to assign two value to the fucntion using tuple unpacking
    
    if marker.upper() == "X":
        return ('X','O')
    else:
        return ('O','X')

'''
this is function to place the marker in position
'''
def place_marker(board,marker,position):
    board[position] = marker

def win_check(board,marker):
    # check with boolean value if there is a win in the game..
    return ((board[7] == marker and board[8] == marker and board[9] == marker) or 
            (board[4] == marker and board[5] == marker and board[6] == marker) or 
            (board[1] == marker and board[2] == marker and board[3] == marker) or 
            (board[1] == marker and board[4] == marker and board[7] == marker) or 
            (board[2] == marker and board[5] == marker and board[8] == marker) or 
            (board[3] == marker and board[6] == marker and board[9] == marker) or 
            (board[1] == marker and board[5] == marker and board[9] == marker) or 
            (board[3] == marker and board[5] == marker and board[7] == marker))

'''
this is a function will randomly choose a player to play first
'''
def random_player():
    rand_player = random.randint(0,1)
    if rand_player == 0:
        return "PLAYER 1"
    else:
        return "PLAYER 2"

'''
A fucntion to return a boolean value if there is an empty space in the board
'''
def space_check(board,position):
    return board[position] == ' '

'''
A function to return a boolean value if the board is full ot not
'''
def full_board_check(board):
    for i in range(1,10):
        if space_check(board,i):
            return False
    # will return true if the board is full..
    return True

'''
A function to control the position if its been used by player
'''
def player_choice(board):
    position = 0
    # a while loop check if the position has been used or not
    while position not in [1,2,3,4,5,6,7,8,9] or not space_check(board,position):
        position = int(input("Please insert a position (1-9)"))
    return position

'''
a function to restart the game 
'''
def replay():
    ask_player = input("Would you like to play another game ? ENTER y or n ").lower()
    if ask_player == 'y':
        return True
    else:
        return False

'''
Main programm!!
'''
while True:
    
    print("WELCOME TO TIC TAC TOE: DEVELOPED BY PHILIP ALTWAL")
    
    the_board = [' ']*10
    
    # ask the player a marker to choose
    
    player_1_marker, player_2_marker = player_input()
    
    # calling the random function to randomly choos which player will go first 
    
    turn = random_player()
    print(turn," PLAY FIRST")
    
    # ask the player to play
    
    ask_player = input("Ready to play ? y or n ?").lower()
    if ask_player == 'y':
        game_on = True
    else:
        game_on = False
    
    
    while game_on:
        
        if turn == 'PLAYER 1':
            
            # show the board to the player
            display_board(the_board)
            
            # add a position on the board
            position = player_choice(the_board)
            
            # place the marker in postion
            place_marker(the_board,player_1_marker,position)
            
            # check for win
            if win_check(the_board,player_1_marker):
                display_board(the_board)
                print("PLAYER 1 WINS !")
                game_on = False
                
            # check for tie    
            elif full_board_check(the_board):
                display_board(the_board)
                print("TIE GAME !!")
                game_on = False
                
            # switch to player 2 if there is no win or tie
            else:
                turn = 'PLAYER 2'
                
        if turn == 'PLAYER 2':
            
            # show the board to the player
            display_board(the_board)
            
            # add a position on the board
            position = player_choice(the_board)
            
            # place the marker in postion
            place_marker(the_board,player_2_marker,position)
            
            # check for win
            if win_check(the_board,player_2_marker):
                display_board(the_board)
                print("PLAYER 2 WINS !")
                game_on = False
                
            # check for tie
            elif full_board_check(the_board):
                display_board(the_board)
                print("TIE GAME !!")
                game_on = False
                
            # switch to player 2 if there is no win or tie    
            else:
                turn = 'PLAYER 1'
        
    # break out from the main program if replay == False...
    if not replay():
        break
```
